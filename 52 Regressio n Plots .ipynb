```python
# The normal imports
import numpy as np
from numpy.random import randn
import pandas as pd

# Import the stats library from numpy
from scipy import stats

# These are the plotting modules adn libraries we'll use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Command so that plots appear in the iPython Notebook
%matplotlib inline
```

```python
# Luckily, Seaborn comes with an example dataset to use as a pandas DataFrame
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.lmplot('total_bill','tip',tips)
```

```python
sns.lmplot('total_bill','tip',tips,
          scatter_kws={'marker':'o', 'color':'indianred'},
          line_kws={'linewidth':1, 'color':'blue'})
```

```python
# WE can also check out higher-order trends
sns.lmplot("total_bill", "tip", tips,order=4,
           scatter_kws={"marker": "o", "color": "indianred"},
           line_kws={"linewidth": 1, "color": "blue"})
```

```python
sns.lmplot('total_bill','tip',tips,fit_reg=False)
```

```python
tips.head()
```

```python
tips['tip_percentage'] = 100 * (tips['tip']/tips['total_bill'])
```

```python
tips.head()
```

```python
sns.lmplot('size','tip_percentage',tips)
```

```python
sns.lmplot('size','tip_percentage',tips, x_jitter=.1)
```

```python
sns.lmplot('size','tip_percentage',tips,x_estimator=np.mean)
```

```python
sns.lmplot('total_bill','tip_percentage',tips,hue='sex',markers=['x','o'])
```

```python
sns.lmplot('total_bill','tip_percentage',tips,hue='day')
```

```python
# Finally it should be noted that Seabron supports LOESS model fitting
url = 'http://en.wikipedia.org/wiki/Local_regression'

sns.lmplot("total_bill", "tip_percentage", tips, lowess=True, line_kws={"color": 'black'});
```

```python
# The lmplot() we've been using is actually using a lower-level function, regplot()
sns.regplot("total_bill","tip_percentage",tips)

```

```python
# reg_plot can be added to existing axes without modifying anything in the figure

# Create figure with 2 subplots
fig, (axis1,axis2) = plt.subplots(1,2,figsize=(15,15),sharey =True)

sns.regplot("total_bill","tip_percentage",tips,ax=axis1)
sns.violinplot(x="size", y="tip_percentage",data=tips, palette="muted")
```
